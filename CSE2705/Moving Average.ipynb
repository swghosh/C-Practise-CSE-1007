# Weekly Sales Data

```python
sales_data = [39, 44, 40, 45, 38, 43, 39]
weeks = list(range(1, len(sales_data) + 1))

{'Week %d' % week: sales_data[index] for index, week in enumerate(weeks)}
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(weeks, sales_data, 'go-')
plt.xlabel('Week')
plt.ylabel('Sales (in 1000 $)')
plt.title('Weekly Sales')
plt.show()
```

# 3-week Moving Average

```python
def moving_avg(weeks, sales_data, moving=3):
    moving_weeks = []
    moving_sales = []

    week = moving - 1
    while week < len(weeks):
        cum_sales = 0
        cum_weeks = 0

        for i in range(moving):
            cum_weeks += weeks[(week - i)]
            cum_sales += sales_data[week - i]

        cum_sales = cum_sales / moving
        cum_weeks = cum_weeks / moving
        moving_sales.append(cum_sales)
        moving_weeks.append(cum_weeks)
        week += 1
    return moving_weeks, moving_sales
```

```python
moving_weeks, moving_sales = moving_avg(weeks, sales_data)
```

```python
plt.plot(weeks, sales_data, 'go-')
plt.plot(moving_weeks, moving_sales, 'bo-')
plt.xlabel('Week')
plt.ylabel('Sales (in 1000 $)')
plt.title('Weekly Sales')
plt.legend(['Weekly', '3-Week Moving Average'])
plt.show()
```

# 4-week Moving Average

```python
moving_weeks3, moving_sales3 = moving_avg(weeks, sales_data, 3)
moving_weeks4, moving_sales4 = moving_avg(weeks, sales_data, 4)
```

```python
plt.plot(weeks, sales_data, 'go-')
plt.plot(moving_weeks3, moving_sales3, 'bo-')
plt.plot(moving_weeks4, moving_sales4, 'ro-')
plt.xlabel('Week')
plt.ylabel('Sales (in 1000 $)')
plt.title('Weekly Sales')
plt.legend(['Weekly', '3-Week Moving Average', '4-Week Moving Average'])
plt.show()
```
