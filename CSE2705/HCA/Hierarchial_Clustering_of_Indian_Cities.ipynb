```python
import pandas as pd
data = pd.read_csv('Distance Matrix.csv')
data = data.set_index('Cities')
data
```

```python
from sklearn.impute import SimpleImputer
pairwise_dist = data.values
pairwise_dist = SimpleImputer(strategy='constant', fill_value=0.0).fit_transform(pairwise_dist)
pairwise_dist
```

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(affinity='precomputed', linkage='single')
agc.fit(pairwise_dist)
```

```python
labels = data.columns.values
labels
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 10))
Z = hierarchy.linkage(pairwise_dist, 'single')
dn = hierarchy.dendrogram(Z, 
                          labels=labels, 
                          orientation='right')
```
