The required packages are imported.

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import seaborn as sns
```

We set the style to `whitegrid` for fancy plots.

```python
sns.set()
sns.set_style("whitegrid")
```

Random data points are generated such that they form 500 samples belonging to 4 clusters.

```python
n_samples = 500
n_features = 2
n_clusters = 4

X, y = make_blobs(n_samples=n_samples, 
                  centers=n_clusters,
                  n_features=2, 
                  random_state=11)
```

The data points are visualised.

```python
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], 
            X[:, 1], 
            cmap='coolwarm',
            edgecolor='black',
            c=y)
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('Data Points')
plt.show()
```

k-Means algorithm is used perform clustering. 

```python
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)
print('Mean squared sum: ', kmeans.inertia_)
```

The minimum and maximum data values are gathered.

```python
x_min, x_max = X.min(), X.max()
```

A meshgrid is generated to draw a region over complete input space.

```python
x_space = np.arange(x_min, x_max, 0.05)
ux1, ux2 = np.meshgrid(x_space, x_space)
u = np.stack([ux1.flatten(), ux2.flatten()]).T
```

The complete input space is visualised to know where each cluster belongs.

```python
clusters = kmeans.predict(u)

plt.figure(figsize=(7, 5))
plt.scatter(u[:, 0], 
            u[:, 1], 
            c=clusters, 
            cmap='Spectral')
plt.scatter(X[:, 0], 
            X[:, 1], 
            c=y,
            cmap='coolwarm',
            marker='o',
            edgecolor='black')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('k-Means Clustering')
plt.show()
```
