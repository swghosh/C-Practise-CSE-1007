```r
m <- ncol(cars) - 1
n <- nrow(cars)
X <- cars[,1]
y <- cars[,2]
```

```r
train_size <- 0.8

n_train <- train_size * n
train_indices <- sample(1:n, n_train, replace = F)
cars_train <- cars[train_indices,]
```

```r
plot(cars_train$speed, cars_train$dist)
```

```r
model <- lm(dist ~ speed, cars_train)
model
```

```r
cars_test <- cars[-train_indices, ]
y_test <- cars_test[,2]
y_predict <- predict(model, cars_test)
```

```r
acc <- which(y_predict == y_test)
matrix(y_predict, y_test)
```
