```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_path = './Figure4.tif'
image = cv2.imread(image_path)
gradients = cv2.Laplacian(image, cv2.CV_32F)

plt.imshow(gradients, cmap=plt.cm.)
plt.axis("off")
plt.show()
```

```python
image.shape, gradients.shape, np.max(gradients)
```
