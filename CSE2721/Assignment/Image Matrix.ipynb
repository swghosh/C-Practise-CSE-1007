Import required libraries.

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

Load image as an RGB numpy matrix.

```python
image = plt.imread('e.jpg')
plt.imshow(image)
plt.title('Original Color Image')
plt.show()
```

```python
image.shape
```

Segregate the 3D matrix (RGB) into 3 individual matrices for R, G, B color intensities and visualise them.

```python
r, g, b = image[:, :, 0], image[:, :, 1], image[:, :, 2] 
```

```python
plt.imshow(r, cmap='Reds')
plt.title("Red")
plt.show()

plt.imshow(g, cmap='Greens')
plt.title("Green")
plt.show()

plt.imshow(b, cmap='Blues')
plt.title("Blue")
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) # convert from RGB to Grayscale
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

Generate a 3x3 fresh image with colors of choice.

```python
new_image = np.array([
                      [[255, 0, 0], [0, 255, 0], [0, 0, 255]], 
                      [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
                      [[0, 0, 0], [255, 255, 255], [0, 0, 0]]
                     ], dtype='uint8') # RGB format
plt.imshow(new_image)
plt.show()
```

Segregate into R, G, B components again and visualise them.

```python
r1, g1, b1 = new_image[:, :, 0], new_image[:, :, 1], new_image[:, :, 2] 
```

```python
plt.imshow(r1, cmap='Reds')
plt.title("Red")
plt.show()

plt.imshow(g1, cmap='Greens')
plt.title("Green")
plt.show()

plt.imshow(b1, cmap='Blues')
plt.title("Blue")
plt.show()
```
