```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20, 10))
image_path = './tomatoes.jpg'

# Original Image
plt.subplot(2, 1, 1)
image = cv2.imread(image_path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title('Original Image')

# Binary Tresholding
plt.subplot(2, 1, 2)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
_, part1 = cv2.threshold(hsv_image, 25, 255, cv2.THRESH_BINARY_INV)
plt.imshow(part1)
plt.axis("off")
plt.title('Applying Binary Tresholding (Inverse) on HSV Image')

plt.show()
```

```python
plt.figure(figsize=(20, 10))

# Original Image
plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title('Original Image')

# Binary Tresholding
plt.subplot(2, 1, 2)
red_image = image[:, :, 2]
part2 = cv2.adaptiveThreshold(red_image, 240, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 13, 10)
plt.imshow(part2, cmap=plt.cm.gray)
plt.axis("off")
plt.title('Applying Adaptive Tresholding on Color Image')

plt.show()
```

```python
# taking individual element product
# of (binary thresholded hsv image) and 
# (adaptive thresholded red image)
seg = part2 * part1[:, :, 0]
plt.imshow(seg, cmap=plt.cm.Reds)
plt.title("Segmentation")
plt.axis("off")
plt.show()
```
