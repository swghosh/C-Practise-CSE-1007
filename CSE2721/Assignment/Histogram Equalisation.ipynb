```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
max_val = 255
def create_histogram(data):
    data = data.flatten() # convert to 1-d array
    counts = np.bincount(data, minlength=max_val + 1)
    plt.bar(np.arange(max_val + 1), counts, color='k')
```

```python
def equalize_and_show(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    contrast_image = cv2.equalizeHist(image)

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap=plt.cm.gray)
    plt.axis("off")
    plt.title('Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(contrast_image, cmap=plt.cm.gray)
    plt.axis("off")
    plt.title('Histogram Equalised Image')
    
    plt.show()

    plt.figure(figsize=(7, 3))
    create_histogram(image)
    plt.title('Histogram of Original Image')
    plt.show()
    
    plt.figure(figsize=(7, 3))
    create_histogram(contrast_image)
    plt.title('Histogram of Image after applying Histogram Equalisation')
    plt.show()
```

```python
equalize_and_show('./Figure1.tif')

print('\n' * 3)

equalize_and_show('./Figure3.tif')
```
