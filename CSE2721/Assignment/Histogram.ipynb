```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_path = './Figure1.tif'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

```python
"""
Function will compute a histogram
from a numpy array by counting 
each integer. Bins of the histogram
is representative of single integer.
"""
max_val = 255
def create_histogram(data):
    data = data.flatten() # convert to 1-d array
    counts = np.bincount(data, minlength=max_val + 1)
    return counts
```

```python
counts = create_histogram(image)
plt.figure(figsize=(12, 3))
plt.bar(np.arange(max_val + 1), counts, color='k')
plt.title('Histogram of Individual Gray Levels')
plt.show()
```

```python
plt.hist(image.flatten(), bins=100, color='y')
plt.title("Histogram (image gray levels grouped into 100)")
plt.show()
```

```python
image_path = './c.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.show()
```

```python
r_counts = create_histogram(image_rgb[:, :, 0])
g_counts = create_histogram(image_rgb[:, :, 1])
b_counts = create_histogram(image_rgb[:, :, 2])
```

```python
plt.figure(figsize=(15, 5))
plt.bar(np.arange(max_val + 1), r_counts, color='r')
plt.bar(np.arange(max_val + 1), g_counts, color='g')
plt.bar(np.arange(max_val + 1), b_counts, color='b')

plt.title('Color (Cumulative) Histogram')
plt.show()
```

```python
sample_image = np.array([
    [2, 1, 0, 1],
    [3, 4, 4, 5],
    [3, 4, 4, 5],
    [1, 3, 0, 1]
], dtype=np.uint8)
max_val = np.max(sample_image)

# L1 normalisation
sample_image_norm = sample_image.astype(np.float32) / max_val

plt.imshow(sample_image_norm, cmap=plt.cm.gray)
plt.show()
```

```python
counts = create_histogram(sample_image)
plt.figure(figsize=(5, 2))
plt.bar(np.arange(max_val + 1), counts, color='k')
plt.title('Histogram of Individual Gray Levels')
plt.show()
```
