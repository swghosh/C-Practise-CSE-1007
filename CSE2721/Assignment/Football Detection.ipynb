```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
template = cv2.imread("template.jpg", 0)
frame = cv2.imread("players.jpg", 0)
```

```python
result = cv2.matchTemplate(frame, template, cv2.TM_CCOEFF_NORMED)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
_ = cv2.circle(result, max_loc, 40, 255, 2)
```

```python
h, w = frame.shape
h, w
```

```python
h1, w1 = h // 2, w // 2
h2, w2 = h - h1, w - w1
h1, w1, h2, w2
```

```python
res = cv2.copyMakeBorder(result, h1, h2, w1, w2, cv2.BORDER_CONSTANT)
res.shape
```

```python
plt.imshow(frame, cmap=plt.cm.gray)
plt.axis("off")
plt.title("Frame")
plt.show()

plt.figure(figsize=(1, 1))
plt.imshow(template, cmap=plt.cm.gray)
plt.axis("off")
plt.title("Template")
plt.show()
```

```python
shape = frame.shape
res = cv2.resize(result, (frame.shape[1], frame.shape[0]))
```

```python
plt.imshow(result, cmap=plt.cm.gray)
plt.axis("off")
plt.show()

plt.imshow(res * 0.5 + frame, cmap=plt.cm.gray)
plt.axis("off")
plt.title("Detected object using Template Matching")
plt.show()
```
