```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10, 10))
image_path = './Figure3.tif'

# Original Image
plt.subplot(2, 2, 1)
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap=plt.cm.gray)
plt.axis("off")
plt.title('Original Image')

# Binary Tresholding
plt.subplot(2, 2, 2)
_, rice = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(rice, cmap=plt.cm.gray)
plt.axis("off")
plt.title('Applying Binary Tresholding')

# Adaptive Tresholding (Gaussian)
plt.subplot(2, 2, 3)
rice = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)
plt.imshow(rice, cmap=plt.cm.gray)
plt.axis("off")
plt.title('Applying Mean Adaptive Tresholding')

# Adaptive Tresholding (Mean)
plt.subplot(2, 2, 4)
rice = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 2)
plt.imshow(rice, cmap=plt.cm.gray)
plt.axis("off")
plt.title('Applying Gaussian Adaptive Tresholding')

plt.show()
```
